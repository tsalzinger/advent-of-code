```kotlin
fun String.toRotation(): Int {
    val sign = when(val direction = this[0]) {
        'L' -> -1
        'R' -> 1
        else -> error("Unexpected dial direction $direction encountered")
    }
    val rotations = substring(1).toInt()

    return sign * rotations
}
```

```kotlin
val DIAL_STARTING_POSITION = 50
val AVAILABLE_DIAL_POSITIONS = 100
```

```kotlin
val rotations = day.input().lines().map { it.toRotation() }
```

```kotlin
val exampleInput = listOf(
    "L68",
    "L30",
    "R48",
    "L5",
    "R60",
    "L55",
    "L1",
    "L99",
    "R14",
    "L82",
)
```

```kotlin
val exampleRotations = exampleInput.map { it.toRotation() }
exampleRotations
```

```kotlin
val performMove = { (position, zeroCount): Pair<Int, Int>, rotation: Int ->
    require(rotation != 0) { "Undefined count semanthics for a 0 rotation" }

    var currentZeroCount = zeroCount
    var currentPosition = position

    val direction = rotation.sign
    val ticks = rotation.absoluteValue

    repeat(ticks) {
        currentPosition += direction
        when {
            currentPosition == -1 -> currentPosition += AVAILABLE_DIAL_POSITIONS
            currentPosition == AVAILABLE_DIAL_POSITIONS -> currentPosition -= AVAILABLE_DIAL_POSITIONS
        }
    }

    if (currentPosition == 0) {
        currentZeroCount++
    }

    currentPosition to currentZeroCount
}

fun List<Int>.debugMoves(): List<Pair<Int, Int>> {
    return runningFold(DIAL_STARTING_POSITION to 0, performMove)
}

fun List<Int>.countMoves() = fold(DIAL_STARTING_POSITION to 0, performMove).second
```

```kotlin
val performRunningCountMove = { (position, zeroCount): Pair<Int, Int>, rotation: Int ->
    require(rotation != 0) { "Undefined count semanthics for a 0 rotation" }

    var currentZeroCount = zeroCount
    var currentPosition = position

    val direction = rotation.sign
    val ticks = rotation.absoluteValue

    repeat(ticks) {
        currentPosition += direction
        when {
            currentPosition == -1 -> currentPosition += AVAILABLE_DIAL_POSITIONS
            currentPosition == AVAILABLE_DIAL_POSITIONS -> currentPosition -= AVAILABLE_DIAL_POSITIONS
        }

        if (currentPosition == 0) currentZeroCount++
    }

    currentPosition to currentZeroCount
}

fun List<Int>.debugRunningCountMoves(): List<Pair<Int, Int>> {
    return runningFold(DIAL_STARTING_POSITION to 0, performRunningCountMove)
}

fun List<Int>.runningCountMoves() = fold(DIAL_STARTING_POSITION to 0, performRunningCountMove).second
```

```kotlin
val part2Solution = rotations.runningCountMoves()
part2Solution
```
