```kotlin
fun String.toLongRange() = split("-").map { it.toLong() }.let { it.first()..it.last() }
```

```kotlin
fun String.toLongRanges() = lines().flatMap { it.split(",") }.map { it.toLong() }
```

```kotlin
val exampleInput = "11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124"
val exampleRanges = exampleInput.toLongRanges()
exampleRanges
```

```kotlin
val inputRanges = input.toLongRanges()
inputRanges
```

```kotlin
fun String.isRepeatingTwice(): Boolean {
    if (count() % 2 != 0) return false

    val half = count() / 2

    return take(half) == takeLast(half)
}

fun Long.isRepeatingTwice() = toString().isRepeatingTwice()
```

```kotlin
fun List<LongRange>.sumOfRepeatingIds(): Long {
    return flatMap { range ->
        range.filter {
            it.isRepeatingTwice()
        }
    }.sum()
}
```

```kotlin
fun String.isSillyPattern(): Boolean {
    val maxRepetitions = count()

    return (2..maxRepetitions).firstOrNull { repetition ->
        if (maxRepetitions % repetition != 0) return@firstOrNull false
        val parts = chunked(maxRepetitions / repetition)

        parts.all { it == parts.first() }
    } != null

}

fun Long.isSillyPattern() = toString().isSillyPattern()
```

```kotlin
listOf(
    "11",
    "111",
    "121",
    "1212",
    "121212",
    "113451134511345",
).map { it.toLong().isSillyPattern() }
    .joinToString("\n")
```

```kotlin
fun List<LongRange>.sumOfSillyPatterns(): Long {
    return flatMap { range ->
        range.filter {
            it.isSillyPattern()
        }
    }.sum()
}
```

```kotlin
exampleRanges.sumOfSillyPatterns()
```

```kotlin
val part2Solution = inputRanges.sumOfSillyPatterns()
part2Solution
```
