```kotlin
fun String.parseInput() = lines().map { bank ->
    bank.map { battery -> "$battery".toInt() }
}
```

```kotlin
fun List<Int>.getMaxJoltage(): Int {
    val first = dropLast(1).max()
    val second = subList(indexOf(first) + 1, count()).max()

    return "$first$second".toInt()
}
```

```kotlin
parsedExampleInput.map { it.getMaxJoltage() }
```

```kotlin
fun List<List<Int>>.getMaxJoltages() = map { it.getMaxJoltage() }
```

```kotlin
fun List<List<Int>>.getSumOfMaxJoltages() = getMaxJoltages().sum()
```

```kotlin
val BATTERY_COUNT = 12
fun List<Int>.getMaxJoltage2(): Long {
    var previous = -1
    var digits = mutableListOf<Int>()
    var remainder = this

    repeat(BATTERY_COUNT) { index ->
        val remainingDigits = BATTERY_COUNT - index
        val searchable = remainder.dropLast(remainingDigits - 1)
        previous = searchable.max()

        digits += previous
        remainder = remainder.drop(remainder.indexOf(previous) + 1)
    }

    return digits.joinToString("").toLong()
}

fun List<List<Int>>.getMaxJoltages2() = map { it.getMaxJoltage2() }
fun List<List<Int>>.getSumOfMaxJoltages2() = getMaxJoltages2().sum()
```
